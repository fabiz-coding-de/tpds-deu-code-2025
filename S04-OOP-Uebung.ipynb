**3.d.** Führe die nächste Zelle aus, um Alben-Daten aus der Datei `alben.csv` in eine Variable `alben_liste` zu laden.

```python
import csv
alben_liste = []
try:
    with open('alben.csv', mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        current_id = 1
        for row in reader:
            album = MusikAlbum(id=current_id, titel=row['titel'],
                kuenstler=row['kuenstler'], medium=row['medium'],
                genre=row['genre'], jahr=int(row['jahr']), preis=float(row['preis'])
            )
            alben_liste.append(album)
            current_id += 1
    print(f"Erfolgreich {len(alben_liste)} Alben geladen.")

except FileNotFoundError:
    print("Die Datei 'alben.csv' wurde nicht gefunden.")
```
